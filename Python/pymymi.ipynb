```python
import pymyami as pma
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from pymyami import calculate_seawater_correction
```

```python
Mg = (0.0132075, 0.026415, 0.05283, 0.079245, 0.10566)

Ca = (0.00259, 0.00518, 0.01036 ,0.01554,0.02072)
SO4 = (0.00706, 0.01412, 0.02824 ,0.04236,0.05648)
T = (10, 20, 30)
S = (25, 35, 50)

pkvals = []

for temp in T:
    for sal in S:
        for i in range(len(SO4)):
            ks = calculate_seawater_correction(TempC=temp, Sal=sal, Mg=0.05283, Ca=0.01036, BOH4=0.00042, SO4=SO4[i], Sr=0)
            pkvals.append([SO4[i], ks['K1'], ks['K2'], temp, sal])
            pkvals_df = pd.DataFrame(pkvals, columns=['SO4_concentration', 'K1_ratio', 'K2_ratio', 'Temperature', 'Salinity'])

pkvals_df.to_csv("pymyami_SO4.csv", index=False)
```
